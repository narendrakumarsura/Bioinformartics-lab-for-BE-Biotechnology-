## Search for Restriction Sites

### Subtask:
Write a Python function to iterate through the defined restriction enzymes and search for their recognition sites within the prepared DNA sequence. The function should return the enzyme name, its recognition site, and all starting positions where it's found.


**Reasoning**:
I will define the `find_restriction_sites` function as described in the instructions, which iterates through the restriction enzymes and their recognition sequences, searching for all occurrences within the DNA sequence. This function will collect and return the enzyme name, recognition site, and all found starting positions.



```python
def find_restriction_sites(dna_sequence, restriction_enzymes):
    found_sites = {}
    for enzyme_name, recognition_sequence in restriction_enzymes.items():
        positions = []
        # Search for all occurrences of the recognition sequence
        i = 0
        while True:
            index = dna_sequence.find(recognition_sequence, i)
            if index == -1:
                break
            positions.append(index)
            i = index + 1  # Move past the found site to find subsequent occurrences

        if positions:
            found_sites[enzyme_name] = {
                'recognition_sequence': recognition_sequence,
                'positions': positions
            }
    return found_sites

# Call the function with the prepared DNA sequence and restriction enzymes
restriction_site_results = find_restriction_sites(dna_sequence, restriction_enzymes)

# Print the results for verification
print("Identified Restriction Sites:")
if restriction_site_results:
    for enzyme, data in restriction_site_results.items():
        print(f"  Enzyme: {enzyme}")
        print(f"    Recognition Sequence: {data['recognition_sequence']}")
        print(f"    Positions: {data['positions']}")
else:
    print("  No restriction sites found for the defined enzymes.")
```
